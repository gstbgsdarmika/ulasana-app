```python
import pandas as pd
from google_play_scraper import app, Sort, reviews

def scrape_reviews(app_id, reviews_per_score, filename="dataset_google-play.csv"):
  # Initialize dictionary for storing reviews
  scraped_reviews = {"reviewId": [], "userName": [], "content": [], "score": []}

  # Loop for each score
  for score in range(1, 6):
    # Scrape reviews for the current score
    reviews_data, _ = reviews(
        app_id,
        lang="id",
        country="id",
        sort=Sort.MOST_RELEVANT,
        count=reviews_per_score,
        filter_score_with=score,
    )

    # Add scraped data to the dictionary
    for review in reviews_data:
      scraped_reviews["reviewId"].append(review["reviewId"])
      scraped_reviews["userName"].append(review["userName"])
      scraped_reviews["content"].append(review["content"])
      scraped_reviews["score"].append(review["score"])

  # Convert dictionary to DataFrame
  app_reviews_df = pd.DataFrame(scraped_reviews)

  # Save the DataFrame to a CSV file
  app_reviews_df.to_csv(filename, index=None, header=True)
```

```python
# Example usage
app_id = "com.binance.cloud.tokocrypto"
reviews_per_score = 1000
filename = "tokocrypto_reviews.csv"

scrape_reviews(app_id, reviews_per_score, filename)
```

```python
df = pd.read_csv("tokocrypto_reviews.csv")

#Count the amount of review data per score value
score_counts = df["score"].value_counts()
print(score_counts)
```
