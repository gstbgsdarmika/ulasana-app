##POS Tag

```python
import string
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer

def tag_part_of_speech(text):
    text_splited = text.split(' ')
    text_splited = [''.join(c for c in s if c not in string.punctuation) for s in text_splited]
    text_splited = [s for s in text_splited if s]
    pos_list = pos_tag(text_splited)
    noun_count = len([w for w in pos_list if w[1] in ('NN','NNP','NNPS','NNS')])
    adjective_count = len([w for w in pos_list if w[1] in ('JJ','JJR','JJS')])
    verb_count = len([w for w in pos_list if w[1] in ('VB','VBD','VBG','VBN','VBP','VBZ')])
    return[noun_count, adjective_count, verb_count]
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
df['nouns'], df['adjectives'], df['verbs'] = zip(*df['preprocessed'].apply(
        lambda comment: tag_part_of_speech(comment)))
df['length'] = df['preprocessed'].apply(lambda x : len(x))
df['nouns_vs_length'] = df['nouns'] / df['length']
df['adjectives_vs_length'] = df['adjectives'] / df['length']
df['verbs_vs_length'] = df['verbs'] /df['length']
df['nouns_vs_words'] = df['nouns'] / df['word_count']
df['adjectives_vs_words'] = df['adjectives'] / df['word_count']
df['verbs_vs_words'] = df['verbs'] / df['word_count']
df["mean_word_len"] = df["preprocessed"].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
```

##Does it has Positive Terms

```python
#Positive Keywords
pos_word = '/content/drive/MyDrive/Penambangan Data Tekstual/Dataset/positive_keyword.txt'

with open(pos_word, 'r') as file:
    pos_words = [line.strip() for line in file]

# Function to count positive words
def count_pos_words(text):
    return sum(word in text for word in pos_words)

# Apply the functions to create new columns
df['pos_word'] = df['preprocessed'].apply(count_pos_words)

# Display the updated DataFrame
df
```

##Does it has Negative Terms

```python
#Negative Keywords
neg_word = '/content/drive/MyDrive/Penambangan Data Tekstual/Dataset/negatif_keyword.txt'

with open(neg_word, 'r') as file:
    neg_words = [line.strip() for line in file]

# Function to count negative words
def count_neg_words(text):
    return sum(word in text for word in neg_words)

# Apply the functions to create new columns
df['neg_word'] = df['preprocessed'].apply(count_neg_words)

# Display the updated DataFrame
df
```

##K-Fold and hyperparameters

```python
from sklearn.model_selection import train_test_split

X = df.drop(['content', 'sentiment', 'preprocessed'], axis=1)

Y = df['sentiment'].values
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=69)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X
```

```python
K.clear_session()

k_folds = 10
kf = KFold(n_splits=k_folds, shuffle=True, random_state=69)

learning_rates = [0.001, 0.0005]
num_nodes_list = [16, 32]
num_layers_list = [1, 2, 3]
epochs_list = [50, 100]

history_list = []
best_model = None
best_val_acc = 0.0

for learning_rate in learning_rates:
    for num_nodes in num_nodes_list:
        for num_layers in num_layers_list:
            for epochs in epochs_list:
                print(f"\nTesting hyperparameters: Learning Rate={learning_rate}, Num Nodes={num_nodes}, Num Layers={num_layers}, Epochs={epochs}")

                val_acc_list = []

                for fold_num, (train_index, val_index) in enumerate(kf.split(X_train)):
                    X_train_fold, X_val_fold = X.iloc[train_index], X.iloc[val_index]
                    y_train_fold, y_val_fold = Y[train_index], Y[val_index]

                    model = tf.keras.Sequential()
                    model.add(Dense(num_nodes, activation='relu', input_shape=(X_train_fold.shape[1],)))
                    model.add(Dropout(0.2))
                    for _ in range(num_layers):
                        model.add(Dense(num_nodes, activation='relu'))
                        model.add(Dropout(0.2))
                    model.add(Dense(1, activation='sigmoid'))

                    loss = tf.keras.losses.BinaryCrossentropy()
                    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
                    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

                    history = model.fit(X_train_fold, y_train_fold, epochs=epochs, batch_size=128, validation_data=(X_val_fold, y_val_fold), verbose=0)

                    val_acc_list.append(np.max(history.history['val_accuracy']))

                    print(f"  Fold {fold_num + 1} Validation Accuracy: {np.max(history.history['val_accuracy'])}")

                average_val_acc = np.mean(val_acc_list)

                print(f"  Average Validation Accuracy Across Folds: {average_val_acc}")

                history_list.append({
                    'learning_rate': learning_rate,
                    'num_nodes': num_nodes,
                    'num_layers': num_layers,
                    'epochs': epochs,
                    'average_val_acc': average_val_acc
                })

                if average_val_acc > best_val_acc:
                    best_val_acc = average_val_acc
                    best_model = model

best_hyperparameters = max(history_list, key=lambda x: x['average_val_acc'])
print("\nBest Hyperparameters:")
print(best_hyperparameters)
model_name_h5 = 'feateng_NN' + str(round(best_hyperparameters['average_val_acc'], 2)) + '_.h5'
best_model.save(model_name_h5)
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
import pickle

X_train, X_test, y_train, y_test = train_test_split(df['preprocessed'], df['sentiment'], test_size=0.2, random_state=69)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df['sentiment'])

k_folds = 10
kf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=69)

learning_rates = [0.001]
num_nodes_list = [32, 64]
num_layers_list = [1, 2]
epochs_list = [50, 100, 200]
max_features_list = [100, 200, 1000, 2500]

history_list = []
best_vocabulary = None
best_model = None
best_val_acc = 0.0

for learning_rate in learning_rates:
    for num_nodes in num_nodes_list:
        for num_layers in num_layers_list:
            for epochs in epochs_list:
                for max_features in max_features_list:
                    print(f"\nTesting hyperparameters: Learning Rate={learning_rate}, Num Nodes={num_nodes}, Num Layers={num_layers}, Epochs={epochs}, Max Features={max_features}")

                    val_acc_list = []

                    tfidf_vectorizer = TfidfVectorizer(max_features=max_features)
                    X_tfidf = tfidf_vectorizer.fit_transform(df['preprocessed'])

                    for fold_num, (train_index, val_index) in enumerate(kf.split(X_tfidf, y_encoded)):
                        X_train_fold, X_val_fold = X_tfidf[train_index], X_tfidf[val_index]
                        y_train_fold, y_val_fold = y_encoded[train_index], y_encoded[val_index]

                        model = tf.keras.Sequential()
                        model.add(Dense(num_nodes, input_shape=(X_train_fold.shape[1],), activation='relu'))
                        model.add(Dropout(0.5))
                        for _ in range(num_layers):
                            model.add(Dense(num_nodes, activation='relu'))
                            model.add(Dropout(0.2))
                        model.add(Dense(1, activation='sigmoid'))

                        optimizer = Adam(learning_rate=learning_rate)
                        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

                        history = model.fit(X_train_fold.toarray(), y_train_fold, epochs=epochs, batch_size=128, validation_data=(X_val_fold.toarray(), y_val_fold), verbose=0)

                        val_acc_list.append(np.max(history.history['val_accuracy']))

                        print(f"  Fold {fold_num + 1} Validation Accuracy: {np.max(history.history['val_accuracy'])}")

                    average_val_acc = np.mean(val_acc_list)

                    print(f"  Average Validation Accuracy Across Folds: {average_val_acc}")

                    history_list.append({
                        'learning_rate': learning_rate,
                        'num_nodes': num_nodes,
                        'num_layers': num_layers,
                        'epochs': epochs,
                        'max_features': max_features,
                        'average_val_acc': average_val_acc
                    })

                    if average_val_acc > best_val_acc:
                        best_val_acc = average_val_acc
                        best_model = model
                        best_vocabulary = tfidf_vectorizer.get_feature_names_out()

best_hyperparameters = max(history_list, key=lambda x: x['average_val_acc'])
print("\nBest Hyperparameters:")
print(best_hyperparameters)

model_name_h5 = f'TF-IDF_model_{round(best_hyperparameters["average_val_acc"], 2)}_{best_hyperparameters["max_features"]}.h5'
best_model.save(model_name_h5)

if best_vocabulary is not None:
    with open('best_tfidf_vocabulary_NN.pkl', 'wb') as f:
        pickle.dump(best_vocabulary, f)
```
